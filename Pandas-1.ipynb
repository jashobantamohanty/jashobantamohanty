# DataFrame
* multiple rows and columns
* similar to Excel

```python
# DataFrame Creation->values, arrays, random nos., list, tuples, dict
df1 = pd.DataFrame([[1,2,3,6],[5,6,7,9],[4,2,8,1]])
df1
```

```python
# Assigning names to rows & columns
df1.index = ['R1', 'R2', 'R3']
df1.columns = ['C1','C2' ,'C3', 'C4']
df1
```

```python
df1.index = 'R11 R22 R33'.split()
df1.columns = 'C11 C22 C33 C44'.split()
df1
```

```python
# Create DF with arrays
df2 = pd.DataFrame(np.array([[1,2,3],[4,5,6]]))
df2
```

```python
# Create DF with random nos, list and tuple (2,3)
df3 = pd.DataFrame(np.random.randint(1,1000,6).reshape(2,3))
df3
```

```python
df4 = pd.DataFrame(np.random.randint(1,2000,15).reshape(5,3))
df4
```

```python
df5 = pd.DataFrame([[1,4,5,6,7],[2,3,6,4,9]])
df5
```

```python
df6 = pd.DataFrame(((1,4,5,6,7),(2,3,6,4,9)))
df6
```

```python
d1 = {'Names':['Deepa','Nithu','Shashu'],'Age':[33,3,6]}
d1
```

```python
df7 = pd.DataFrame(d1)
df7
```

```python
# Create DF with shape 5,6 with any values and give names to rows & cols
df8 = pd.DataFrame(data=np.random.randint(1,2000,30).reshape(5,6),index='R1 R2 R3 R4 R5'.split(),columns='A B C D E G'.split())
df8
```

```python
#Attributes-> values,index,columns,ndim,size,shape
print(df8.ndim)
print(df8.size)
print(df8.shape)
print(df8.index)
print(df8.columns)
print(df8.values)
```

```python
df8
```

```python
# Indexing and slicing-> iloc,loc and conditions
#iloc-> index based location
df8.iloc[1:5,1:3]
```

```python
# loc-> location(names of col & rows)
df8.loc[:,['A','C','E']]
```

```python
df8.loc[['R1','R3'],['A','C','E']]
```

```python
#condition based retrievel
df8[df8>1000]
```

```python
df8>1000
```
